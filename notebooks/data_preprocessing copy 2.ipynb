```python
import pandas as pd
```

```python
# Read in the data
df = pd.read_csv('../data/retail_sales.csv')
df
```

```python
num_products = df['Product ID'].nunique()
print(f"There are {num_products} unique products in the dataset.")
```

```python
# Convert date columns to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Drop irrelevant columns
df.drop(['Row ID', 'Product ID', 'Order ID', 'Customer ID', 'Customer Name', 'Postal Code', 'Ship Date', 'Ship Mode', 'Segment', 'Country', 'City'], axis=1, inplace=True)
```

```python
df.duplicated()
```

```python
df
```

```python

# Group by 'Category', 'Sub-Category' and 'Order Date', then count the number of sales
df = df.groupby(['Category', 'Sub-Category', 'Order Date']).size().reset_index(name='Sales Count')

# Display the result
df
```

```python
from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder object
le = LabelEncoder()

# Fit and transform the 'Sub-Category' column
df['Sub-Category'] = le.fit_transform(df['Sub-Category'])

df['Category'] = le.fit_transform(df['Category'])
```

```python
df
```

```python
train = df[df['Order Date'].dt.year.isin([2015, 2016, 2017])]
test = df[df['Order Date'].dt.year == 2018]
```

```python
train.to_csv('../data/processed_data/train_data.csv', index=False)
test.to_csv('../data/processed_data/test_data.csv', index=False)
```
